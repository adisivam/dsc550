##### 7. Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.

```python
## Create a subset of dataframe using only Publisher and Global Sales columns and use groupby, 
## sum and sort values to show the highest global sales (sum) by publisher

video_df[['Publisher', 'Global_Sales']].groupby(['Publisher']).sum().sort_values(by='Global_Sales', ascending=False).head(n=5)
```

##### 10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
## Trying to find out the median without converting NaN values to check if we are getting error

video_df["User_Score"].median()
```

As expected, the median calculation has been failed with TypeError as the data has some non float values. So, we will try to calculate the median after converting "tbd" values into NAN

```python
#### Replace the "tbd" value into nan and assign it to different dataframe

median_df = video_df.replace('tbd', np.nan)
median_df.head(10)
```

```python
## Calculating the median value after converting tbd to nan values

median_value = median_df["User_Score"].median()
print("The median values for user score: {}".format(median_value))
```

```python
## Replace all NaN entries in the user score column with the median value.

median_df['User_Score'].fillna(value=median_df['User_Score'].median(), inplace=True)
median_df.head(10)
```

```python
## Calculating the number of rows and columns present in the dataframe

print("Number of rows and columns present in the dataframe: {}".format(median_df.shape))
```

```python
## Checking if the User Score has any NaN values

cnt_records = median_df["User_Score"].isna().sum()
print("Count of nan records present in median dataframe: {}".format(cnt_records))
```
